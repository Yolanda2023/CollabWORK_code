```python
def save_data(data, path, filename):

    if not os.path.exists(path):
        os.makedirs(path)

    dataframe = pd.DataFrame(data)
    dataframe.to_csv(path + filename, encoding='utf_8_sig', mode='a', index=False, sep=',', header=False )
#     print('成功')
```

```python
rolldown_time = 0
id_timestamp_set = set()
while rolldown_time <= 10:
    print('rolldown_time = ',rolldown_time)
    length = len(browser.find_elements_by_xpath('//div[@data-testid="cellInnerDiv"]'))
    print('length = ',length)
    for i in range(length):
        try:
            publisher_id = browser.find_element_by_xpath('//div[@data-testid="cellInnerDiv"]['+str(i+1)+']//div[@class="css-1dbjc4n"]/article//div[@data-testid="User-Name"]/div[2]//a/div[@dir="ltr"]/span').text

            content_web_list = browser.find_elements_by_xpath('//div[@data-testid="cellInnerDiv"]['+str(i+1)+']//div[@class="css-1dbjc4n"]//div[@dir="auto"]/span')
            content_list = [content.text for content in content_web_list]
            content = ''.join(content_list)

            publish_time = browser.find_element_by_xpath('//div[@data-testid="cellInnerDiv"]['+str(i+1)+']//div[@class="css-1dbjc4n"]/article//time').get_attribute('datetime')

            id_timestamp = publisher_id + publish_time
#             print("**********")
#             print(publisher_id)
#             print(content)
#             print(id_timestamp)
            if id_timestamp not in id_timestamp_set:
                id_timestamp_set.add(id_timestamp)
                save_data([[publisher_id,content,publish_time]], path, filename)
        except:
            continue
    
    # roll down
    browser.execute_script("document.documentElement.scrollTop="+str(rolldown_time*2000))
    rolldown_time += 1
    sleep(random.random()*6+1)

    
```
